## 3) 10 부터 49까지의 값을 가지는 벡터를 만드세요.

```
b = np.arange(10,50)
b
```

## 4) 위 3)에서 만든 벡터의 순서를 바꾸세요.

요건 여러가지 방법이 있습니다.

```
# 방법 1.

%time print(b[::-1])
```

```
# 방법 2.

%time print(np.flipud(b))
```

```
# 방법 3.

%time print(np.ascontiguousarray(b[::-1]))
```

방법 1이 이해하기 가장 어렵지만 가장 빠릅니다.

파이썬의 list slicing을 응용한 방법으로써,

> **[(시작위치) : (끝위치) : (stepsize)]** 

로 이해하시면 되겠습니다

더 이해가 필요한 분은 [여기](https://stackoverflow.com/questions/15748001/reversed-array-in-numpy)를 보세요!

## 6) 벡터 [1,2,0,0,4,0] 에서 원소의 값이 0이 아닌 원소만 선택한 벡터를 만드세요.

```
x = np.array([1,2,0,0,4,0])

%time x[x > 0]
print(x[x > 0])

%time x[x != 0]
print(x[x != 0])
```
