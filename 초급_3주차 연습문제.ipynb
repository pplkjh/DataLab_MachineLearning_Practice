#### 파일 parsing을 위한 google api

```python
# ------------- 데이터 입출력을 위한 google api 호출 -------------

# for google autorization
from google.colab import auth
auth.authenticate_user()
from googleapiclient.discovery import build
drive_service = build('drive', 'v3')


# download data to python
file_id = '1kuYZbsez3RrpQm6Fxh0dS-KGJCzuLxr5'                                           # file id를 넣어야 합니다. 여기서는 제가 넣어놔서 바로 실행하면 됩니다. https://drive.google.com/open?id=1mdaxfa161UhvgNGKTtmC0gFS1EJpXDru 를 참고
import io
from googleapiclient.http import MediaIoBaseDownload
request = drive_service.files().get_media(fileId=file_id)
downloaded = io.BytesIO()                                                               # 여기 라인에서 보시다시피, 파일을 가져오는데 성공하면 downloaded 변수안에 담도록 되어있습니다.
downloader = MediaIoBaseDownload(downloaded, request)
done = False
while done is False:
  _, done = downloader.next_chunk()  
downloaded.seek(0)

print('complete')
```

#### pandas로 읽어와서 dataframe으로 만들기

```python
# 자 그럼 위 결과에서 score칼럼만 선택면되겠죠?

df.groupby('gender')['score'].mean()           # gender를 기준으로 groupby하고, score칼럼의 평균을 구하라
                                               # 결과의 데이터타입은 뭘까요? 칼럼이 하나뿐이기때문에 series가 됩니다
```
