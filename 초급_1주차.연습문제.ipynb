## 연습문제 2

- 구구단을 9단까지 모두 print 하시오

print했을때 결과가 다음과 같은 형태로 나오면 성공!!!<br>

2<br>
4<br>
6<br>
...<br>
18<br>
3<br>
6<br>
9<br>
....<br>


```
num_range = range(1,10)   # 정수 1~9 까지의 list를 range를 이용하여 만들고

for i in num_range[1:]:   # 첫번째 루프에서는 숫자중 2~9 까지만 선택(슬라이싱)
  for j in num_range:     # 두번째 루프에서는 숫자 1~9 까지 모두 선택
    print (i * j)         # 첫번째 루프 돌때의 각 숫자마다  아래 루프의 각 숫자가 실행되면서 그 두개를 각각 곱하게 됩니다. identation에 주의!!!
```

위에코드 잘 되긴 하는데.

수업시간에 배우셨던걸 좀더 응용하시면 아래와 같이 만들수도 있습니다.

```
num_range = range(1,10)
for i in num_range[1:]:
  print(i, '단')
  for j in num_range:
    print ('\t', i, ' x ', j, ' = ', i * j)
```

## 연습문제 3

- 1~100까지 자연수 중에, 홀수만 모두 하나씩 출력하시오

<br>

위 연습문제 1에서 했던걸 응용하면 됩니다. 크게 두가지 방법이 잇습니다.

### 방법 1.

```
for i in range(1,101):
  if not i % 2 == 0:              # if를 if not으로 해버린다면? true가 아니고 false일때만 실행되겠죠?
    print (i)
```

### 방법 2.

```
for i in range(1,101):
  if i % 2 != 0:              # if는 그대로두고 == 을 !=으로 바꾸면 됩니다. !=은 ==의 반대. 즉 같지 않다를 의미합니다. 2로 나눠서 0이 아니면 홀수죠 뭐
    print (i)
```

## 연습문제 4.

- numpy를 쓰지 않고 1부터 100까지의 평균을 구해보시오

<br>

수업시간에 잠깐 보여드렸었지만 연습삼아 직접 해보시면 좋습니다.

일단 평균을 구하는 방법과 순서를 생각해 봅시다. 그렇게 그대로 코드를 만들면 되니까.

그 순서는 다음과 같습니다.

1. 주어진 모든 숫자를 다 더한다.

2. 총 몇개인지 주어진 숫자의 갯수를 구한다

3. 위 1을 위 2로 나눈다

해봅시다.

```
# 먼저 1~100 까지 숫자를 다 더해야겠죠?

num = range(1,101)                  # 더이상 말이 必要韓家????

sum_value = 0                       # 초기값을 0으로 만들어서 변수를 하나 만들어놓습니다.. 왜 필요한지는 아래 코드를 보면서 잘 생각해봅시다

for i in num:                       # 1 ~ 100을 돌면서
  
  sum_value = sum_value + i         # 여기가 핵심입니다. sum_value는 for문안에서 자기자신 값에다가 i를 더하네요.
                                    # 첫번째인 1이 들어오면 sum_value는 초기값 + 1 이 되니까 0 + 1 해서 1이 됩니다. for문돌기전에 초기값을 만들어놧거든요. 0으로.
                                    # 두번째인 2가 들어오면 현재 sum_value값은?? 0이아닌 1이 되있습니다. 그러면 1 + 2 = 3
                                    # 세번째인 3이 들어오면 2 + 3 = 5. 이런식으로 100까지 계속 반복..
                                    # 즉 자기자신을 계속 i와 더하면서 업데이트 시켜 나가겠네요.

sum_value                           # for문을 다 돌고나면? sum_value변수에는 1~100까지의 모든 값의 합이 담겨 있겠네요
```

```
# 합친값을 나눌 갯수를 구한다. 간단하다. 앞서 배운 len 이용하면된다.

len(num)
```

```
# 평균을 구하자. 더한걸 갯수로 나누면 되니까.

sum_value / len(num)
```
